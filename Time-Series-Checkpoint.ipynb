```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
passenger_data = pd.read_csv('AirPassengers.csv', parse_dates=[0])
```

```python
passenger_data['month'] = pd.to_datetime(passenger_data['Month']).dt.month
```

```python
passenger_data['year'] = pd.to_datetime(passenger_data['Month']).dt.year
```

```python
passenger_data['day'] = pd.to_datetime(passenger_data['Month']).dt.day
```

```python
passenger_data.tail()
```

```python
passenger_data.rename(columns = {'Month' : 'Date', '#Passengers' : 'Passengers'}, inplace = True)
```

```python
passenger_data
```

```python
plt.figure(figsize=(16,5))
plt.plot(passenger_data['Date'],passenger_data['Passengers'])
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.title('Passengers over Date')
plt.show()
```

```python
plt.figure(figsize=(16,7))
plt.plot(passenger_data['month'],passenger_data['Passengers'])
plt.xlabel('month')
plt.ylabel('Passengers')
plt.title('Passengers over month')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(passenger_data['year'],passenger_data['Passengers'])
plt.xlabel('year')
plt.ylabel('Passengers')
plt.title('Passengers over year')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(passenger_data['day'],passenger_data['Passengers'])
plt.xlabel('day')
plt.ylabel('Passengers')
plt.title('Passengers over day')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20,7), dpi= 80)
sb.boxplot(x='year', y='Passengers', data=passenger_data, ax=axes[0])
sb.boxplot(x='month', y='Passengers', data=passenger_data.loc[~passenger_data.year.isin([1949, 1960]), :])

# Set Title
axes[0].set_title('Year-wise Box Plot\n(The Trend)', fontsize=18); 
axes[1].set_title('Month-wise Box Plot\n(The Seasonality)', fontsize=18)
plt.show()
```
